```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
vx = np.fromfile("./NYX_f32/data/VelocityX.dat", dtype=np.float32)
vy = np.fromfile("./NYX_f32/data/VelocityY.dat", dtype=np.float32)
vz = np.fromfile("./NYX_f32/data/VelocityZ.dat", dtype=np.float32)
vtot = (vx**2 + vy**2 + vz**2)
```

```python
vtot_5 = np.where(vtot > 0, 1 / vtot**0.5, 0)
vtot_4 = np.where(vtot > 0, 1 / vtot**0.4, 0)
vtot_3 = np.where(vtot > 0, 1 / vtot**0.3, 0)
vtot_2 = np.where(vtot > 0, 1 / vtot**0.2, 0)
vtot_1 = np.where(vtot > 0, 1 / vtot**0.1, 0)
```

```python
def to_weights(weights, max_weight=4):
    max_val = np.max(np.abs(weights))

    abs_weights = np.abs(weights)
    log_weights = np.log2(max_val / abs_weights, where=abs_weights != 0)  # 避免对 0 求 log2
    log_weights = np.floor(log_weights).astype(np.int32)  # 转为整数，模拟 C++ 行为

    # 归一化权重的逻辑
    int_weights = np.where(abs_weights == 0, 0, max_weight - log_weights)
    int_weights = np.where(log_weights > max_weight, 0, int_weights)
            
    return int_weights
```

```python
def weight_hist(weights):
    arr = [0, 0, 0, 0, 0]
    for weight in weights:
        arr[weight] += 1
    for i in range(5):
        print(f"weight = {i}: {arr[i]}")
```

```python
vtots = [vtot_5, vtot_4, vtot_3, vtot_2, vtot_1]
labels = ["$v_{tot}^{0.5}$", "$v_{tot}^{0.4}$", "$v_{tot}^{0.3}$", "$v_{tot}^{0.2}$", "$v_{tot}^{0.1}$"]
colors = ['yellow', 'blue', 'green', 'purple', 'orange']

fig, axes = plt.subplots(5, 2, figsize=(15, 20), sharex='col', sharey=False)

for i, (vtot, label, color) in enumerate(zip(vtots, labels, colors)):
    # histogram of vtot
    axes[i, 0].hist(vtot, bins=200, range=(0,2), color=color, alpha=0.7, edgecolor='black')
    axes[i, 0].set_title(f"Histogram of {label} (Vtot)", fontsize=14)
    axes[i, 0].set_ylabel("Frequency", fontsize=12)
    axes[i, 0].grid(True, linestyle='--', alpha=0.6)

    # histogram of weight
    weights = to_weights(vtot)
    print(f"weights of vtot_{5 - i}")
    weight_hist(weights)
    axes[i, 1].hist(weights, bins=5, color=color, alpha=0.7, edgecolor='black')
    axes[i, 1].set_title(f"Histogram of {label} (Weights)", fontsize=14)
    axes[i, 1].grid(True, linestyle='--', alpha=0.6)

for ax in axes[-1, :]:
    ax.set_xlabel("$v_{tot}$ or Weights", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
from _collections import defaultdict
import matplotlib.pyplot as plt

# 文件和数据集设置
files = ["Dummy_compression_ratios.txt", "SZ3_compression_ratios.txt", "PMGARD_compression_ratios.txt"]
compressors = ["Dummy", "SZ3", "PMGARD"]  # 压缩器名称
datasets = ["Hurricane", "NYX", "GE_small"]  # 数据集名称
baseline_file = open("/Users/wenboli/uky/qoi-control/build/Result/CR.txt", 'r').readlines()
error_bound = []
baselines = defaultdict(list)
for line in baseline_file[:20]:
    error_bound.append(float(line.rstrip().split("Error Bound: ")[1].split(",")[0]))

# 每个数据集对应的最佳参数
best_weights = [7, 7, 5]
best_block_sizes = [4, 4, 64]

# 初始化存储数据结构
best_curves = defaultdict(lambda: defaultdict(list))  # 存储最佳曲线
baselines = defaultdict(lambda: defaultdict(list))  # 存储 baseline 曲线

# 读取并解析数据
for file, compressor in zip(files, compressors):
    with open(f"./build/Result/{file}", "r") as f:
        lines = f.readlines()
    for line in lines:
        ds = line.rstrip().split(", ")[1]
        if ds not in datasets:
            continue
        if line.startswith("BP"):  # Baseline 数据
            bitrate = 32 / float(line.rstrip().split("cr = ")[1])
            baselines[ds][compressor].append(bitrate)
        else:  # 其他数据
            max_weight = int(line.rstrip().split("max_weight: ")[1].split(", ")[0])
            block_size = int(line.rstrip().split("block_size: ")[1].split(", ")[0])
            bitrate = 32 / float(line.rstrip().split("cr = ")[1])

            # 提取最佳参数曲线
            for idx, dataset in enumerate(datasets):
                if ds == dataset and max_weight == best_weights[idx] and block_size == best_block_sizes[idx]:
                    best_curves[dataset][compressor].append(bitrate)

# 压缩器对应的颜色
compressor_colors = {"Dummy": "blue", "SZ3": "green", "PMGARD": "red"}

# 绘制最佳曲线比较图
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)  # 每个数据集一个子图
for i, dataset in enumerate(datasets):
    ax = axs[i]
    for compressor in compressors:
        color = compressor_colors[compressor]  # 获取压缩器对应的颜色

        # 绘制 baseline 曲线
        baseline_x = baselines[dataset][compressor]  # X 轴为 baseline 比特率
        ax.plot(baseline_x, error_bound, label=f"{compressor} Baseline", linestyle='--', marker='.', color=color, alpha=0.8)
        
        # 绘制最佳参数的曲线
        best_x = best_curves[dataset][compressor]  # X 轴为比特率
        ax.plot(best_x, error_bound, label=f"{compressor}, Max Weight {best_weights[i]}, Block Size {best_block_sizes[i]}", linestyle='-', marker='o', color=color, alpha=1)
    
    # 子图设置
    ax.set_title(f"Best Curve Comparison - {dataset}")
    ax.set_xlabel("Bitrate")
    if i == 0:
        ax.set_ylabel("Relative Error")
    ax.set_yscale('log')  # 设置 Y 轴为对数刻度
    ax.grid(False)
    ax.legend()

# 调整布局并显示
plt.tight_layout()
plt.show()
```

```python
from _collections import defaultdict
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 数据处理 - VTOT
baseline_file = open("/Users/wenboli/uky/qoi-control/build/Result/CR.txt", 'r').readlines()
error_bounds = []
baselines_vtot = defaultdict(list)
for line in baseline_file[:20]:
    error_bounds.append(float(line.rstrip().split("Error Bound: ")[1].split(",")[0]))

for line in baseline_file:
    qoi = line.rstrip().split("QoI: ")[1].split(",")[0]
    method = line.rstrip().split("Method: ")[1].split(",")[0]
    if method == "PMGARD":
        method = "PMGARD-HB"
    if method == "SZ3_delta":
        method = "PSZ3-delta"
    eb = float(line.rstrip().split("Error Bound: ")[1].split(",")[0])
    cr = float(line.rstrip().split("cr = ")[1].split(",")[0])
    bitrate = 64 / cr
    if qoi == "Vtot" and eb in error_bounds:
        baselines_vtot[method].append(bitrate)

for method in baselines_vtot.keys():
    baselines_vtot[method].sort()

VTOT = open("/Users/wenboli/uky/test/ProDM/build/Result/SZ3_compression_ratios_d64.txt", 'r').readlines()
best_curves_vtot = []
for line in VTOT:
    data_name = line.rstrip().split(", ")[1]
    if data_name == "GE_small" and line.startswith("WBP"):
        max_weight = int(line.rstrip().split("max_weight: ")[1].split(",")[0])
        block_size = int(line.rstrip().split("block_size: ")[1].split(",")[0])
        eb = float(line.rstrip().split("Error Bound: ")[1].split(",")[0])
        cr = float(line.rstrip().split("cr = ")[1].split(",")[0])
        bitrate = 64 / cr
        if max_weight == 5 and block_size == 64 and eb in error_bounds:
            best_curves_vtot.append(bitrate)

# 数据处理 - T
baselines_t = defaultdict(list)
for line in baseline_file:
    qoi = line.rstrip().split("QoI: ")[1].split(",")[0]
    method = line.rstrip().split("Method: ")[1].split(",")[0]
    if method == "PMGARD":
        method = "PMGARD-HB"
    if method == "SZ3_delta":
        method = "PSZ3-delta"
    eb = float(line.rstrip().split("Error Bound: ")[1].split(",")[0])
    cr = float(line.rstrip().split("cr = ")[1].split(",")[0])
    bitrate = 64 / cr
    if qoi == "T" and eb in error_bounds:
        baselines_t[method].append(bitrate)

for method in baselines_t.keys():
    baselines_t[method].sort()

T = open("/Users/wenboli/uky/test/ProDM/build/Result/T_compression_ratios_d64.txt", 'r').readlines()
best_curves_t = []
for line in T:
    compressor = line.rstrip().split("Compressor: ")[1].split(",")[0]
    if line.startswith("WBP"):
        max_weight = int(line.rstrip().split("max_weight: ")[1].split(",")[0])
        block_size = int(line.rstrip().split("block_size: ")[1].split(",")[0])
        eb = float(line.rstrip().split("Error Bound: ")[1].split(",")[0])
        cr = float(line.rstrip().split("cr = ")[1].split(",")[0])
        bitrate = 64 / cr
        if compressor == "SZ3" and max_weight == 5 and block_size == 64 and eb in error_bounds:
            best_curves_t.append(bitrate)

method_colors = {"PMGARD-HB": "red", "PSZ3-delta": "blue", "Best": "green"}

# 绘制子图
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# 子图 1: VTOT
axes[0].set_title("VTOT")
for method in baselines_vtot.keys():
    axes[0].plot(baselines_vtot[method], error_bounds, label=f"{method}",
                 linestyle='--', marker='.', color=method_colors[method], alpha=0.8)
axes[0].plot(best_curves_vtot, error_bounds, label="Best",
             linestyle='-', marker='.', color=method_colors["Best"], alpha=1)
axes[0].legend(loc="lower left", fontsize=10, frameon=True)
axes[0].set_xlabel("Bitrate", fontsize=12)
axes[0].set_ylabel("Relative QoI Error", fontsize=12)
axes[0].set_yscale('log')
axes[0].yaxis.set_major_locator(ticker.LogLocator(base=10.0, numticks=7))
axes[0].yaxis.set_minor_locator(ticker.NullLocator())
axes[0].yaxis.set_major_formatter(ticker.LogFormatterMathtext(base=10))

# 子图 2: T
axes[1].set_title("T")
for method in baselines_t.keys():
    axes[1].plot(baselines_t[method], error_bounds, label=f"{method}",
                 linestyle='--', marker='.', color=method_colors[method], alpha=0.8)
axes[1].plot(best_curves_t, error_bounds, label="Best",
             linestyle='-', marker='.', color=method_colors["Best"], alpha=1)
axes[1].legend(loc="lower left", fontsize=10, frameon=True)
axes[1].set_xlabel("Bitrate", fontsize=12)
axes[1].set_ylabel("Relative QoI Error", fontsize=12)
axes[1].set_yscale('log')
axes[1].yaxis.set_major_locator(ticker.LogLocator(base=10.0, numticks=7))
axes[1].yaxis.set_minor_locator(ticker.NullLocator())
axes[1].yaxis.set_major_formatter(ticker.LogFormatterMathtext(base=10))

# 调整布局
plt.tight_layout()
plt.show()
```
